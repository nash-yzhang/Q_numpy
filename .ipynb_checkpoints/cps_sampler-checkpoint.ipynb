```python
import numpy as np
import Q_num as qn
import importlib
importlib.reload(qn);
import time
```

```python
currentspeed = 0
pnt_num = 100
energy_thre = max(1/pnt_num,0.0065)
maxiter = 500
repulf = .01
rstf   = 1.5
iterN  = 0
energy = 1
sph_pnt = np.random.rand(pnt_num,3)-.5
sph_pntQ = qn.qn(sph_pnt)
```

```python
while energy > energy_thre and iterN < maxiter:
    diff_vec = np.squeeze(sph_pntQ[np.newaxis,:] - sph_pntQ[:,np.newaxis])
    vec_dist = np.squeeze(np.arcsin(diff_vec.norm()/2)*2)
    proj_diff_vec = (diff_vec+qn.reflect(sph_pntQ,diff_vec))/2;
    vecForce = qn.nansum(np.squeeze(1/vec_dist[...,None]**2*np.squeeze(proj_diff_vec.normalize())))
    vecForce *= repulf/pnt_num
    vecForce = np.squeeze(vecForce)
    currentspeed += vecForce
    rotaxis = np.squeeze((sph_pntQ*np.squeeze(currentspeed)).imag())
    rotaxis = np.squeeze(qn.exp(vecForce.norm()[...,None]/2*np.squeeze(rotaxis.normalize())))
    # rotaxis[np.isnan(rotaxis.norm())] = qn.qn([0,0,0,1])
    sph_pntQ = rotaxis*sph_pntQ*rotaxis.conj()
    sph_pntQ = np.squeeze(sph_pntQ.normalize())
    currentspeed /= rstf
    currentspeed = np.squeeze(currentspeed)
    energy_all = np.sort(vec_dist)
    energy = np.std(energy_all[1,:])
    iterN += 1
```
